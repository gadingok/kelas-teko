 adalah:

 memberikan representation deret dari nilai bilangan 
. Buat program Python ringkas yang menghitung jumlah parsial hingga 
 untuk mengaproksimasi:

```python
def hitung_pi_per_4(N):
    
    hasil = 0
    for n in range(N + 1):
        hasil += (-1)**n / (2*n + 1)
    return hasil

N = 5000
aproksimasi_pi_per_4 = hitung_pi_per_4(N)

print(f"Aproksimasi pi/4 dengan N = {N}: {aproksimasi_pi_per_4}")
```

3. Buat suatu fungsi yang dinamakan arctan_taylor yang mengambil parameter masukan (input) x dan N. Fungsi ini kemudian memberikan keluaran (output) polinomial Taylor derajat 
 dari fungsi 
 yang dievaluasi pada x:
 
Selanjutnya, ujilah fungsi tersebut untuk suatu nilai 
 dan 
. Catatan uji kewarasan ("sanity check"): hasilnya harus konsisten dengan soal no. 2.

jawab

```python
def arctan_taylor(x, N):
    
    hasil = 0
    for k in range(N + 1):
        hasil += (-1)**k * (x**(2*k + 1)) / (2*k + 1)
    return hasil

x = 1
N = 5000
hasil_taylor = arctan_taylor(x, N)

print(f"Hasil arctan_taylor({x}, {N}) = {hasil_taylor}")

import math

print(f"Nilai arctan(1) dari math library = {math.atan(1)}")
```

4. Buatlah suatu fungsi bernama sqrt_integral yang mengambil parameter masukan u, p and N dan memberikan jumlahan Riemann (Rieman sum) dengan menggunakan titik-titik tengah (midpoints) 
 dari suatu partisi berukuran 
:
 
 
Di sini 
 dan 
 untuk titik-titik ujung 
.

 jawab

```python
def sqrt_integral(u, p, N):
    
    delta_x = u / N
    hasil = 0
    for k in range(1, N + 1):
        x_k = k * delta_x
        x_k_minus_1 = (k - 1) * delta_x
        x_k_star = (x_k + x_k_minus_1) / 2
        hasil += (1 / (1 + x_k_star**p)**0.5) * delta_x
    return hasil

u = 1.0
p = 2.0
N = 1000

aproksimasi_integral = sqrt_integral(u, p, N)
print(f"Aproksimasi integral untuk u={u}, p={p}, N={N}: {aproksimasi_integral}")
```

5. Nyatakanlah matriks persegi berukuran 2x2 secara umum sebagai suatu "list of lists". Sebagai contoh kasus khusus, matriks
 
jika dituliskan sebagai "list of lists" adalah [[2,-1],[5,7]]. Kita dapat mengganti angka-angka di dalamnya dengan variabel Python secara umum yang dapat berupa input dari pengguna (user). Selanjutnya, tulislah fungsi yang dinamakan inverse_a yang mengambil parameter masukan a dengan nilai tertentu dan memberikan keluaran "list of lists" untuk invers dari matriks di bawah ini:

jawab


```python
def inverse_a(a):
    
    determinan = -1 - a**2
    if determinan == 0:
        return "Matriks tidak memiliki invers (determinan = 0)"
    else:
        return [
            [-1 / determinan, -a / determinan],
            [-a / determinan, 1 / determinan]
        ]

# Contoh penggunaan
a = 2
invers = inverse_a(a)
print(f"Invers dari matriks [[1, {a}], [{a}, -1]] adalah: {invers}")

a = 0
invers = inverse_a(a)
print(f"Invers dari matriks [[1, {a}], [{a}, -1]] adalah: {invers}")
```

6. buat suatu fungsi yang dinamakan root_sequence yang mengambil parameter-parameter masukan 
 dan 
, keduanya bilangan bulat positif, dan fungsi ini kemudian memberikan suku 
 (nilai 
 ke-
) dari suatu barisan bilangan dengan ketentuan:
 
Apakah barisan ini akan konvergen ke nilai-nilai yang berbeda untuk nilai awal 
 yang berbeda?

jawab


```python
def root_sequence(a, N):

    x = a
    for n in range(N):
        x = 1 + x**0.5
    return x

a = 2
N = 10
hasil = root_sequence(a, N)
print(f"x_{N} untuk a={a}: {hasil}")

a_values = [1, 2, 5, 10, 100]
N_values = [10, 50, 100]

for a in a_values:
    print(f"\nAnalisis konvergensi untuk a = {a}:")
    for N in N_values:
        hasil = root_sequence(a, N)
        print(f"  x_{N} = {hasil}")
```

7. Buat suatu fungsi bernama fib_less_than yang mengambil satu masukan 
 dan mengembalikan keluaran berupa list dari bilangan Fibonacci yang nilainya kurang dari 
. Setelah itu, lanjutkan dengan membuat fungsi bernama fibonacci_primes yang mengambil parameter masukan 
 dan mengembalikan keluaran berupa list dari bilangan Fibonacci yang kurang dari 
 sekaligus juga sebagai bilangan prima.

 jawab
 

```python
def fib_less_than(N):

    
    fib_list = []
    a, b = 0, 1
    while a < N:
        fib_list.append(a)
        a, b = b, a + b
    return fib_list

N = 50
fib_numbers = fib_less_than(N)
print(f"Bilangan Fibonacci kurang dari {N}: {fib_numbers}")

def is_prime(n):

    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def fibonacci_primes(N):
   
    fib_numbers = fib_less_than(N)
    prime_fib_list = [num for num in fib_numbers if is_prime(num)]
    return prime_fib_list


N = 100
fib_primes = fibonacci_primes(N)
print(f"Bilangan Fibonacci prima kurang dari {N}: {fib_primes}")
```

9. Dengan menggunakan pustaka Matplotlib, buatlah plot fungsi-fungsi berikut ini:

jawab 

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.linspace(-9, 9, 400)
y1 = np.sqrt(np.abs(x1))

plt.figure()
plt.plot(x1, y1)
plt.title("f(x) = sqrt(|x|)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()

x2 = np.linspace(0, 4 * np.pi, 400)
y2 = np.sin(x2) + np.sin(2 * x2)

plt.figure()
plt.plot(x2, y2)
plt.title("f(x) = sin(x) + sin(2x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()

x3 = np.linspace(-1, 1, 400)
y3 = np.zeros_like(x3)  
for i, x in enumerate(x3):
    if x != 0:
        y3[i] = 2 * x * np.log(np.abs(x))

plt.figure()
plt.plot(x3, y3)
plt.title("f(x) = 2xln|x|")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()

x4 = np.linspace(-2, 3, 400)
y4 = (x4 + 2) * (x4 + 1) * (x4 - 1) * (x4 - 2) * (x4 - 3)

plt.figure()
plt.plot(x4, y4)
plt.title("f(x) = (x+2)(x+1)(x-1)(x-2)(x-3)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()

x5 = np.linspace(-2, 2, 400)
y5 = np.exp(-x5**2)

plt.figure()
plt.plot(x5, y5)
plt.title("f(x) = e^(-x^2)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```

10. Buatlah suatu fungsi bernama parametric_plots yang mengambil parameter masukan a dan k yang kemudian menghasilkan keluaran plot persamaan parametrik:
 
untuk 
. Sertakan judul untuk setiap subplot untuk menampilkan nilai-nilai 
 dan 
. Gunakan plt.axis('equal') agar tampilan gambar proporsional.

jawab

```python
import matplotlib.pyplot as plt
import numpy as np

def parametric_plots(a, k):
    
    t = np.linspace(0, 2 * np.pi, 400)
    x = 2 * k * np.cos(t) - a * np.cos(k * t)
    y = 2 * k * np.sin(t) - a * np.sin(k * t)

    plt.figure()
    plt.plot(x, y)
    plt.title(f"a = {a}, k = {k}")
    plt.axis('equal')
    plt.show()

a_values = [3, 5, 7]
k_values = [2, 3, 4]

for a in a_values:
    for k in k_values:
        parametric_plots(a, k)
```
