# Analisis Pengaruh Koefisien Drag terhadap Kecepatan Maksimum Mobil Balap

# BAB I: PENDAHULUAN

### 1.1 Latar Belakang

Dalam dunia otomotif, khususnya pada mobil balap, kecepatan dan akselerasi merupakan dua aspek penting yang menentukan performa kendaraan. Salah satu faktor utama yang memengaruhi akselerasi adalah gaya hambat udara (aerodynamic drag). Gaya ini dipengaruhi oleh koefisien drag ($C_d$), yaitu parameter yang menunjukkan seberapa besar hambatan udara yang diterima kendaraan berdasarkan bentuk aerodinamisnya.

Semakin besar nilai $C_d$, maka semakin besar pula gaya hambat yang akan memperlambat percepatan kendaraan. Untuk memahami efek langsung dari variasi $C_d$ terhadap performa awal mobil balap, perlu dilakukan pendekatan kuantitatif berbasis simulasi.

Namun, karena sistem gerak tersebut menghasilkan persamaan diferensial non-linier yang sulit diselesaikan secara analitik, maka pendekatan numerik diperlukan. Dalam tugas besar ini, digunakan dua metode numerik, yaitu *Euler* dan *Runge-Kutta orde 4 (RK4)*, untuk menyelesaikan model gerak mobil berdasarkan hukum Newton II.

Studi ini akan menyimulasikan perubahan kecepatan dua mobil balap dengan koefisien drag berbeda ($C_d = 0{,}4$ dan $C_d = 0{,}7$) dalam rentang waktu *1 hingga 20 detik*. Dengan pendekatan ini, diharapkan dapat diperoleh pemahaman yang lebih baik mengenai pengaruh $C_d$ terhadap laju akselerasi awal kendaraan.

# BAB III: MODEL MATEMATIS

### 3.1 Asumsi dan Pendekatan Fisik

Untuk menyederhanakan pemodelan sistem gerak mobil terhadap gaya hambat udara, digunakan beberapa asumsi sebagai berikut:

- Mobil bergerak lurus di atas lintasan datar.
- Gaya mesin ($F_{\text{mesin}}$) dianggap konstan.
- Gaya gesek ban dan rolling resistance diabaikan.
- Gaya hambat udara merupakan satu-satunya gaya resistif.
- Perubahan massa mobil diabaikan (massa tetap).
- Simulasi hanya dilakukan dalam interval waktu 0 hingga 20 detik.

---

### 3.2 Gaya Hambat Udara

Gaya hambat udara dinyatakan dengan persamaan:

$$
F_d = \frac{1}{2} \rho A C_d v^2
$$

Dengan:
- $\rho$: densitas udara (kg/m³)
- $A$: luas penampang depan mobil (m²)
- $C_d$: koefisien drag
- $v$: kecepatan mobil (m/s)

---

### 3.3 Persamaan Gerak

Berdasarkan hukum Newton II, resultan gaya pada mobil:

$$
m \cdot \frac{dv}{dt} = F_{\text{mesin}} - F_d
$$

Substitusi gaya hambat menghasilkan:

$$
\frac{dv}{dt} = \frac{F_{\text{mesin}}}{m} - \frac{1}{2m} \rho A C_d v^2
$$

Definisikan konstanta:

$$
a = \frac{F_{\text{mesin}}}{m}, \quad b = \frac{1}{2m} \rho A C_d
$$

Sehingga model matematis utama:

$$
\frac{dv}{dt} = a - b v^2
$$

Persamaan tersebut merupakan ODE non-linier yang menunjukkan perubahan kecepatan terhadap waktu, dan akan diselesaikan secara numerik.

---

### 3.4 Tujuan Model

Model ini digunakan untuk:
- Menyimulasikan grafik kecepatan terhadap waktu $v(t)$ dalam 20 detik pertama.
- Membandingkan pengaruh dua nilai $C_d$ terhadap percepatan awal mobil.
- Membandingkan akurasi dan tren hasil metode Euler dan Runge-Kutta orde 4 (RK4).

# 4. Metodologi

### 4.1 Metode Euler

Metode Euler adalah metode numerik eksplisit yang digunakan untuk menyelesaikan persamaan diferensial biasa (ODE). Berdasarkan persamaan gerak yang telah dirumuskan sebelumnya:

$$
\frac{dv}{dt} = a - b v^2
$$

Metode Euler memperkirakan nilai kecepatan pada waktu selanjutnya menggunakan rumus:

$$
v_{i+1} = v_i + \Delta t \cdot (a - b v_i^2)
$$

Dengan:
- $v_i$: kecepatan saat ini
- $v_{i+1}$: kecepatan setelah selang waktu $\Delta t$
- $a = \frac{F_{\text{mesin}}}{m}$
- $b = \frac{1}{2m} \rho A C_d$

Metode ini digunakan untuk mensimulasikan kecepatan mobil selama 20 detik pertama dengan interval waktu 0.1 detik.

### 4.2 Parameter Simulasi

Berikut adalah parameter fisika dan numerik yang digunakan:

| Parameter       | Simbol | Nilai       | Satuan      |
|----------------|--------|-------------|-------------|
| Massa mobil    | $m$    | 900         | kg          |
| Gaya mesin     | $F$    | 4000        | N           |
| Densitas udara | $\rho$ | 1.225       | kg/m³       |
| Luas frontal   | $A$    | 1.8         | m²          |
| Langkah waktu  | $\Delta t$ | 0.1     | s           |
| Waktu simulasi | $t_{\text{max}}$ | 30 | s       |
| Cd Mobil A     | $C_d$  | 0.4         | –           |
| Cd Mobil B     | $C_d$  | 0.7         | –           |

```python
import matplotlib.pyplot as plt 

def euler(Cd):

    m = 900       # Massa mobil (kg)
    F_mesin = 4000  # Gaya mesin (N)
    rho = 1.225   # Massa jenis udara (kg/m^3)
    A = 1.8       # Luas penampang depan mobil (m^2)
    dt = 0.1      # Ukuran langkah waktu (detik)
    t_max = 30    # Waktu simulasi maksimum (detik)

    # a adalah percepatan akibat gaya mesin
    a = F_mesin / m
    # b adalah konstanta yang terkait dengan hambatan udara
    b = 0.5 * rho * A * Cd / m
    
    t = [0]
    v = [0]
    
    while t[-1] < t_max:
        v_new = v[-1] + dt * (a - b * v[-1]**2)
        v.append(v_new)
        t.append(t[-1] + dt)
    return t, v

# Fungsi untuk menampilkan tabel kecepatan
def tampilkan_tabel_euler_1_20():
    print("=== Tabel Kecepatan - Metode Euler (1–20 detik) ===")
    print("{:<8} {:>20} {:>20}".format("Waktu", "Mobil A (Cd=0.4)", "Mobil B (Cd=0.7)"))
    print("-" * 52)

    # Hitung kecepatan untuk Mobil A (Cd=0.4) dan Mobil B (Cd=0.7)
    t_A, v_A = euler(Cd=0.4)
    t_B, v_B = euler(Cd=0.7)

    for i in range(10, 201):  # index ke-10 sampai ke-200 (1.0 s s/d 20.0 s)
        if i % 10 == 0:  # tampilkan tiap 1 detik (setiap 10 langkah dt)
            print(f"{t_A[i]:>6.1f} s {v_A[i]:>18.2f} m/s {v_B[i]:>19.2f} m/s")

# Panggil fungsi untuk menampilkan tabel
tampilkan_tabel_euler_1_20()

# Hitung kembali data untuk plotting (bisa juga menggunakan data dari fungsi tabel jika diinginkan)
t_A, v_A = euler(Cd=0.4)
t_B, v_B = euler(Cd=0.7)

# Buat plot menggunakan Matplotlib
plt.figure(figsize=(10, 5)) # Ukuran figure (lebar, tinggi)
plt.plot(t_A[10:201], v_A[10:201], 'y--', label='Euler - Mobil A (Cd = 0.4)') # Plot Mobil A
plt.plot(t_B[10:201], v_B[10:201], 'g--', label='Euler - Mobil B (Cd = 0.7)') # Plot Mobil B

# Tambahkan label dan judul
plt.xlabel("Waktu (s)")
plt.ylabel("Kecepatan (m/s)")
plt.title("Euler Method: Kecepatan Mobil A vs B (1–20 detik)")

# Tampilkan grid dan legenda
plt.grid(True)
plt.legend()

# Tampilkan plot
plt.show()
```

### 4.5 Metode Runge-Kutta Orde 4 (RK4)

Metode RK4 merupakan metode numerik yang lebih akurat dibandingkan metode Euler. Dengan menggunakan empat perkiraan kemiringan (slope) per langkah, RK4 memperkirakan nilai fungsi pada waktu berikutnya dengan rumus:

$$
\begin{aligned}
k_1 &= f(v_i) \\
k_2 &= f\left(v_i + \frac{1}{2} \Delta t \cdot k_1\right) \\
k_3 &= f\left(v_i + \frac{1}{2} \Delta t \cdot k_2\right) \\
k_4 &= f\left(v_i + \Delta t \cdot k_3\right) \\
v_{i+1} &= v_i + \frac{\Delta t}{6}(k_1 + 2k_2 + 2k_3 + k_4)
\end{aligned}
$$


Dengan $f(v) = a - b v^2$
$$

```python
def rk4(Cd):
    m = 900
    F_mesin = 4000
    rho = 1.225
    A = 1.8
    dt = 0.1
    t_max = 30
    a = F_mesin / m
    b = 0.5 * rho * A * Cd / m
    
    t = [0]
    v = [0]
    while t[-1] < t_max:
        def f(v): return a - b * v**2
        k1 = f(v[-1])
        k2 = f(v[-1] + 0.5 * dt * k1)
        k3 = f(v[-1] + 0.5 * dt * k2)
        k4 = f(v[-1] + dt * k3)
        v_new = v[-1] + (dt / 6) * (k1 + 2*k2 + 2*k3 + k4)
        v.append(v_new)
        t.append(t[-1] + dt)
    return t, v

def tampilkan_tabel_rk4_1_20():
    print("=== Tabel Kecepatan - Metode RK4 (1–20 detik) ===")
    print("{:<8} {:>20} {:>20}".format("Waktu", "Mobil A (Cd=0.4)", "Mobil B (Cd=0.7)"))
    print("-" * 52)

    t_A, v_A = rk4(Cd=0.4)
    t_B, v_B = rk4(Cd=0.7)

    for i in range(10, 201):  # 1.0 s hingga 20.0 s
        if i % 10 == 0:
            print(f"{t_A[i]:>6.1f} s {v_A[i]:>18.2f} m/s {v_B[i]:>19.2f} m/s")

tampilkan_tabel_rk4_1_20()

t_A, v_A = rk4(Cd=0.4)
t_B, v_B = rk4(Cd=0.7)

plt.figure(figsize=(10, 5))
plt.plot(t_A[10:201], v_A[10:201], 'b-', label='RK4 - Mobil A (Cd = 0.4)')
plt.plot(t_B[10:201], v_B[10:201], 'r-', label='RK4 - Mobil B (Cd = 0.7)')
plt.xlabel("Waktu (s)")
plt.ylabel("Kecepatan (m/s)")
plt.title("RK4 Method: Kecepatan Mobil A vs B (1–20 detik)")
plt.grid(True)
plt.legend()
plt.show()
```

### 5.3 Validasi Metode Numerik Berdasarkan ODE (metode matematis)

Untuk memvalidasi hasil metode Euler dan RK4, digunakan pendekatan langsung dari persamaan ODE yang telah diturunkan:

$$
\frac{dv}{dt} = a - b v^2
$$

Dengan:

- $a = \frac{F_{\text{mesin}}}{m} = \frac{4000}{900} = 4.444$
- $b = \frac{1}{2m} \cdot \rho \cdot A \cdot C_d$
- $\rho = 1.225$ kg/m³, $A = 1.8$ m²

Kita akan menghitung nilai percepatan dan kecepatan secara manual (1 langkah) pada $t = 10$ detik dan $t = 20$ detik, lalu membandingkan hasilnya dengan metode Euler dan RK4.

---

##### Mobil A ($C_d = 0.4$)

$$
b = \frac{0.5 \cdot 1.225 \cdot 1.8 \cdot 0.4}{900} = 0.00049
$$

##### Pada $t = 10$ detik

Dari simulasi, kecepatan mendekati:

- Euler: $31.4$ m/s
- RK4  : $31.7$ m/s

Ambil rata-rata: $v = 31.5$ m/s

Hitung percepatan:

$$
\frac{dv}{dt} = 4.444 - 0.00049 \cdot (31.5)^2 \approx 4.444 - 0.486 = 3.958
$$


#### Pada $t = 20$ detik

Ambil $v = 35.1$ m/s (rata-rata dari Euler dan RK4)

$$
\frac{dv}{dt} = 4.444 - 0.00049 \cdot (35.1)^2 \approx 4.444 - 0.604 = 3.84
$$


---

##### Mobil B ($C_d = 0.7$)

$$
b = \frac{0.5 \cdot 1.225 \cdot 1.8 \cdot 0.7}{900} = 0.000857
$$

##### Pada $t = 10$ detik

- Euler: $28.9$ m/s
- RK4  : $29.5$ m/s → ambil rata-rata: $v = 29.2$ m/s

$$
\frac{dv}{dt} = 4.444 - 0.000857 \cdot (29.2)^2 \approx 4.444 - 0.730 = 3.714
$$


#### Pada $t = 20$ detik

- Euler: $32.0$ m/s
- RK4  : $32.9$ m/s → ambil rata-rata: $v = 32.45$ m/s

$$
\frac{dv}{dt} = 4.444 - 0.000857 \cdot (32.45)^2 \approx 4.444 - 0.905 = 3.539
$$

### 5.4 Perbandingan Metode Euler vs RK4

| Aspek                  | Euler                      | RK4                        |
|------------------------|----------------------------|----------------------------|
| Tingkat akurasi        | Rendah - sedang            | Tinggi                     |
| Stabilitas             | Kurang stabil untuk $dt$ besar | Stabil bahkan pada $dt$ kecil |

Kesimpulannya, *metode RK4 lebih akurat dan konsisten*, walaupun metode Euler tetap layak digunakan untuk estimasi awal atau simulasi ringan.

### 5.5 Analisis Aerodinamika dan Kecepatan Maksimum

Koefisien drag ($C_d$) sangat berpengaruh terhadap performa kendaraan, terutama terhadap kecepatan maksimum.

Karena gaya hambat udara meningkat kuadratik terhadap kecepatan ($F_d \propto v^2$), maka mobil dengan $C_d$ lebih tinggi akan lebih cepat mencapai titik di mana gaya dorong sama dengan gaya hambat (steady-state).

> *Mobil A (Cd = 0.4)* mampu mencapai kecepatan maksimum lebih tinggi dibandingkan *Mobil B (Cd = 0.7)* karena hambatan udaranya lebih rendah.
